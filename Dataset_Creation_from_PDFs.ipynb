Description :

This notebook is responsible for creating the raw dataset used in the project. It extracts textual content from official Indian Railways PDF documents such as the Citizen Charter and refund rules. The extracted text is structured into a unified dataset format that serves as the foundational knowledge source for the RAG pipeline.

Key tasks performed:

Load official Indian Railways PDF documents

Extract text page-wise from PDFs

Store extracted content in a structured format (JSON/text)

```python
!pip install pypdf
```

```python
from pypdf import PdfReader
import os
import json

PDF_FOLDER = "Data/PDFs"

documents = []

for file in os.listdir(PDF_FOLDER):
    if file.lower().endswith(".pdf"):
        file_path = os.path.join(PDF_FOLDER, file)
        reader = PdfReader(file_path)

        full_text = ""
        for page in reader.pages:
            text = page.extract_text()
            if text:
                full_text += text + "\n"

        documents.append({
            "source": file,
            "content": full_text.strip()
        })

print("Total documents:", len(documents))
```

```python
with open("Data/railway_rules_raw.json", "w", encoding="utf-8") as f:
    json.dump(documents, f, ensure_ascii=False, indent=2)

print("Dataset saved successfully")
```
